```python
#https://docs.greatexpectations.io/docs/core/introduction/try_gx
#Version for demo 1.0.0

#Prequisites: Python version 3.8 to 3.11

#pip install pandas
#pip install great_expectations


#1- Import the following libraries
import great_expectations as gx
# Pandas -> 
import pandas as pd

# -> Line to check gx core version: print(gx.__version__)

# 2- Download and read the sample data into a Pandas DataFrame.
df = pd.read_csv(
    "./data/RE_technical_merged_2.csv"
)

#3- A Data Context object serves as the entrypoint for interacting with GX components.
context = gx.get_context()

#4- Connect to data and create a Batch.
# Define a Data Source, Data Asset, Batch Definition, and Batch. The Pandas DataFrame is provided to the Batch Definition at runtime to create the Batch.

data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="pd dataframe asset")

batch_definition = data_asset.add_batch_definition_whole_dataframe("batch definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})


#Create an Expectation.
#Expectations are a fundamental component of GX. They allow you to explicitly define the state to which your data should conform.
#Run the following code to define an Expectation that the contents of the column passenger_count consist of values ranging from 2 to 6:
expectation = gx.expectations.ExpectColumnValuesToBeBetween(
    column="Capacity (MW)", min_value=0, max_value=9
)

#Run and get the results!
validation_result = batch.validate(expectation)

print(validation_result)
```

```python
# Import necessary libraries
import great_expectations as gx
import pandas as pd

# Create a sample DataFrame
data = {
    "id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, 30, 35, 40],
    "salary": [50000, 60000, 70000, 80000],
}
df = pd.DataFrame(data)

# Step 1: Create a Great Expectations context
context = gx.get_context()

# Step 2: Add a Pandas Data Source
data_source = context.data_sources.add_pandas(name="pandas_data_source")

# Step 3: Add a Data Asset
data_asset = data_source.add_dataframe_asset(name="sample_dataframe")

# Step 4: Define a Batch (whole DataFrame for Pandas)
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch_definition")

# Step 5: Create a Batch and associate the DataFrame
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})

# Step 6: Create an Expectation Suite
suite_name = "example_suite"
expectation_suite = context.create_expectation_suite(suite_name=suite_name)

# Add Expectations
expectation_suite.add_expectation(
    expectation=batch.expect_column_values_to_not_be_null(column="id")
)
expectation_suite.add_expectation(
    expectation=batch.expect_column_values_to_be_between(column="age", min_value=20, max_value=50)
)
expectation_suite.add_expectation(
    expectation=batch.expect_column_mean_to_be_between(column="salary", min_value=50000, max_value=100000)
)

# Step 7: Validate the Batch against the Expectation Suite
validation_results = batch.validate(expectation_suite)

# Step 8: Print Validation Results
print(validation_results)

# Optional: Save the Expectation Suite
context.save_expectation_suite(expectation_suite, suite_name)
```
